```python
df = pd.get_dummies(df, columns=['Continent', 'Development_level'], drop_first=True)
```

### Model Training 

```python
from sklearn.model_selection import train_test_split

x = df.drop(['Countries', 'Average_CDR'], axis=1)
y = df['Average_CDR']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

### Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

modelLR = LinearRegression()
modelLR.fit(x_train, y_train)

modelLR.score(x_test, y_test)

y_pred_LR = modelLR.predict(x_test)
mse_LR = mean_squared_error(y_test, y_pred_LR)
mae_LR = mean_absolute_error(y_test, y_pred_LR)
print(f'MSE: {mse_LR}')
print(f'MAE: {mae_LR}')
print(f'R^2: {modelLR.score(x_test, y_test)}')
print(f"AVG 'Average_CDR': {df['Average_CDR'].mean()}")
```

### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

modelRF = RandomForestRegressor()
modelRF.fit(x_train, y_train)

y_pred_RF = modelRF.predict(x_test)
mse_RF = mean_squared_error(y_test, y_pred_RF)
mae_RF = mean_absolute_error(y_test, y_pred_RF)
print(f'MSE: {mse_RF}')
print(f'MAE: {mae_RF}')
print(f'R^2: {modelRF.score(x_test, y_test)}')
print(f"AVG 'Average_CDR': {df['Average_CDR'].mean()}")
```

### Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

modelGB = GradientBoostingRegressor(random_state=42)
modelGB.fit(x_train, y_train)

y_pred_GB = modelGB.predict(x_test)
mse_GB = mean_squared_error(y_test, y_pred_GB)
mae_GB = mean_absolute_error(y_test, y_pred_GB)
print(f'MSE: {mse_GB}')
print(f'MAE: {mae_GB}')
print(f'R^2: {modelGB.score(x_test, y_test)}')
print(f"AVG 'Average_CDR': {df['Average_CDR'].mean()}")
```

### XGBoost

```python
import xgboost as xgb
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

modelXGB = XGBRegressor()
modelXGB.fit(x_train, y_train)

y_pred_XGB = modelXGB.predict(x_test)
mse_XGB = mean_squared_error(y_test, y_pred_XGB)
mae_XGB = mean_absolute_error(y_test, y_pred_XGB)
print(f'MSE: {mse_XGB}')
print(f'MAE: {mae_XGB}')
print(f'R^2: {modelXGB.score(x_test, y_test)}')
print(f"AVG 'Average_CDR': {df['Average_CDR'].mean()}")
```
